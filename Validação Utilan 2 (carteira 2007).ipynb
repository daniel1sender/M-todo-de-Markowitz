```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
```

```python
ações = ['VALE5.SA', 'USIM5.SA', 'PETR3.SA', 'CSNA3.SA', 'BBAS3.SA', 'CMIG4.SA', 'CESP6.SA']
carteira = pd.DataFrame()
for i in ações:
    carteira[i] = pdr.DataReader(i, data_source='yahoo', start='2008-03-01', end='2008-03-29')['Adj Close']
```

```python
carteira.head(30)
```

```python
retorno = (carteira/carteira.shift(1))-1
retorno 
```

```python
peso = np.array([0.373, 0.0775, 0.1860, 0.0113, 0.0328, 0.2458, 0.0736])
peso.sum()
```

```python
retorno_mensal = retorno.mean() * 20
retorno_mensal
```

```python
np.dot(retorno_mensal, peso) * 100 
```
